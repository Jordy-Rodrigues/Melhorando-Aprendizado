# ALGORITMO NEURAL_NETWORK.MLPCLASSIFIER
## Classificador de perceptron de várias camadas

```python
# IMPORTAR A BIBLIOTECAS
import numpy as np
from sklearn import datasets
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
# DEFINIR QUE OUTRAS PESSOAS AO APLICAR ESSE MODELO TENHA OS MESMO RESULTADOS
np.random.seed(5)
```

```python
# IMPORTAR O DADASET DISPONÍVEL NO SKLEARN PARA ESSE ESTUDO
iris = datasets.load_iris()
# MOSTRAR A BASE DE DADOS
print(iris)
```

```python
# SEPARAR A VARIÁVEL DEPENDENTE E INDEPENDENTE
x = iris.data
y = iris.target
```

```python
# CRIAR UMA VARIÁVEL COM 150 ÍNDICES 
sufind = np.arange(150)
# EMBARALHAR ESSES VALORES ALEATORIAMENTE
np.random.shuffle(sufind)
```

```python
# OBS: O DATASET CONTEM UM TOTAL DE 150 LINHAS
# PEGAR AS 100 PRIMEIRAS ALEATÓRIAS PARA USAR NA SEPARAÇÃO DOS DADOS DE TREINAMENTO E TESTE
x_treino = x [sufind[ :100], :]
x_teste = x [sufind[100: ], :]
y_treino = y [sufind[ :100]]
y_teste = y [sufind[100: ]]
```

```python
# CRIAR A INTELIGÊNCIA ARTIFICIAL, E PASSAR ALGUNS PARÂMETROS
CLF = MLPClassifier(solver='sgd', hidden_layer_sizes=(3, ), learning_rate_init=0.001, activation='logistic', max_iter=1500, random_state=1)

# 'sgd' ESTAMOS USANDO O STOCHASTIC GRADIENT DESCENT OU EM PORTUGUÊS DESCIDA DO GRADIENTE ESTOCÁSTICO
# hidden_layer_sizes, É A QUANTIDADE DE NEURÔNIOS NA CAMADA ESCONDIDA
# learning_rate_init, TAXA DE APRENDIZADO INICIAL
# activation, FUNÇÃO DE ATIVAÇÃO PARA A CAMADA OCULTA, 'logistic', FUNÇÃO SIGMÓIDE LOGÍSTICA
# max_iter, NÚMERO MÁXIMO DE ITERAÇÕES
# random_state, DETERMINA A GERAÇÃO DE NÚMEROS ALEATÓRIOS
```

```python
# TREINAR O MODELO
CLF.fit(x_treino, y_treino)
# FAZER A PREVISÃO PASSANDO PARA O MODELO DADOS QUE ELE AINDA NÃO VIU
y_pred = CLF.predict_proba(x_teste)
# MOSTRAR 
y_pred

# MATRIZ COM 50 LINHAS E 3 COLUNAS
```

```python
# PROCURAR QUAL COLUNA TEM MAIOR PROBABILIDADE
y_aux = np.argmax(y_pred, 1)
# MOSTRAR A PORCENTAGEM DE ACERTO DO MODELO
accuracy_score(y_teste, y_aux)
# OBS: ACURÁCIA FOI MUITO BAIXA, OBJETIVO MELHORAR ESSA MÉTRICA
```

# NORMALIZANDO OS DADOS

```python
# IMPORTO O MODELO PADRÃO DE NORMALIZAÇÃO
scaler = StandardScaler()

np.random.seed(5)
iris = datasets.load_iris()
x = iris.data
y = iris.target
# TREINAR 
scaler.fit(x)
# FAZER A NORMALIZAÇÃO DOS DADOS
x = scaler.transform(x)

sufind = np.arange(150)
np.random.shuffle(sufind)
x_treino = x [sufind[ :100], :]
x_teste = x [sufind[100: ], :]

y_treino = y [sufind[ :100]]
y_teste = y [sufind[100: ]]
CLF = MLPClassifier(solver='sgd', hidden_layer_sizes=(3, ), learning_rate_init=0.001, activation='logistic', max_iter=1500, random_state=1)
CLF.fit(x_treino, y_treino)
y_pred = CLF.predict_proba(x_teste)
y_aux = np.argmax(y_pred, 1)
# MOSTRAR A ACURÁCIA NOVAMENTE
accuracy_score(y_teste, y_aux)

# OBS: COM A NORMALIZAÇÃO MELHOROU E MUITO A ACURÁCIA DO MODELO PARA 86%
```

# AUMENTAR A TAXA DE APRENDIZADO USANDO NOVOS PARÂMETROS

```python
scaler = StandardScaler()

np.random.seed(5)
iris = datasets.load_iris()
x = iris.data
y = iris.target

scaler.fit(x)
x = scaler.transform(x)

sufind = np.arange(150)
np.random.shuffle(sufind)
x_treino = x [sufind[ :100], :]
x_teste = x [sufind[100: ], :]

y_treino = y [sufind[ :100]]
y_teste = y [sufind[100: ]]

# FAZER UMAS MODIFICAÇÕES NOS PARÂMETROS PARA MELHORAR O MODELO AINDA MAIS
# MODIFICAR A TAXA DE APRENDIZADO INICIAL DE '0.001', PARA '0.01'.
CLF = MLPClassifier(solver='sgd', hidden_layer_sizes=(3, ), learning_rate_init=0.01, activation='logistic', max_iter=1500, random_state=1)
CLF.fit(x_treino, y_treino)
y_pred = CLF.predict_proba(x_teste)
y_aux = np.argmax(y_pred, 1)
accuracy_score(y_teste, y_aux)
# COM ESSE AJUSTE NO MODELO A ACURÁCIA ATINGE SEU MÁXIMO DE 98% E ATINGE SEU LIMITE.
```
